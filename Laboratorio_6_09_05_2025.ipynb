**Gr√°fico #1: Lineal**

```python
import matplotlib.pyplot as plt

x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [1, 3, 5, 3, 1, 3, 5, 3, 1]
y2 = [2, 4, 6, 4, 2, 4, 6, 4, 2]
plt.plot(x, y1, label="line I")
plt.plot(x, y2, label="line H")
plt.plot()

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Graph Example")
plt.legend()
plt.show()
```

**Gr√°fico #2: Barras**

```python
import matplotlib.pyplot as plt

x1 = [1, 3, 4, 5, 6, 7, 9]
y1 = [4, 7, 2, 4, 7, 8, 3]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 6, 2, 6, 2]

plt.show()

plt.bar(x1, y1, label="Cyan", color='c')
plt.bar(x2, y2, label="Amarillo", color='y')
plt.plot()

plt.xlabel("bar number")
plt.ylabel("bar height")
plt.title("Bar Chart Example")
plt.legend()
plt.show()
```

**Gr√°fico #3: Histogramas**

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.random.randn(1000)

m = [m for m in range(len(n))]
plt.bar(m, n)
plt.title("Datos de calificaciones")
plt.show()

plt.hist(n, bins=20)
plt.title("Histograma")
plt.show()
```

**Gr√°fico #4: Marcas**

```python
import matplotlib.pyplot as plt

x1 = [2, 3, 4]
y1 = [5, 5, 5]

x2 = [1, 2, 3, 4, 5]
y2 = [2, 3, 2, 3, 4]
y3 = [6, 8, 7, 8, 7]

plt.scatter(x1, y1)
plt.scatter(x2, y2, marker="*", color='b')
plt.scatter(x2, y3, marker='P', color='0.8')
plt.title('Scatter Plot Example')
plt.show()
```

**Parte 2**

```python
import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_csv("casasboston.csv")
df = datos[["RM","CRIM", "MEDV", "TOWN", "CHAS"]]

df = datos.rename(columns={
	"TOWN":"CIUDAD",
	"CRIM":"INDICE_CRIMEN",
	"INDUS":"PCT_ZONA_INDUSTRIAL",
	"CHAS":"RIO_CHARLES",
	"RM":"N_HABITACIONES_MEDIO",
	"MEDV":"VALOR_MEDIANO",
	"LSTAT":"PCT_CLASE_BAJA"})

df.N_HABITACIONES_MEDIO.plot.hist()
plt.show()
```

**Gr√°fico #6: Dispersi√≥n**

```python
df.plot.scatter(x="INDICE_CRIMEN", y="PCT_CLASE_BAJA", alpha=0.5)
plt.show()
```

**Gr√°fico #7: Barras horizontales**

El m√©todo head con valor 10, se refiere  a mostrar la cantidad de elementos horizontales a mostrar.

```python
valor_por_ciudad = df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
valor_por_ciudad.head(10).plot.barh()
plt.show()
```

**Gr√°fico #8: Pie o de sectores**

```python
df.RIO_CHARLES.value_counts().plot.pie(
    legend=True,
    autopct="%1.1f%%")
plt.show()
```
